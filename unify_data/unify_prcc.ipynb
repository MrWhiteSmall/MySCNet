```python
import os,shutil
from os.path import join as osj
from tqdm import tqdm
```

```python
for oridir,tardir in zip(['train', 'val','processed'], ['train', 'test','processed']):
    dir = f'/root/datasets/prcc/{oridir}'
    tar_dir = f'/root/datasets/prcc_my/{tardir}'
    print(f'Processing {oridir} to {tardir}')
    # break
```

```python
for oridir,tardir in zip(['train', 'val','processed'], ['train', 'test','processed']):
    dir = f'/root/datasets/prcc/{oridir}'
    tar_dir = f'/root/datasets/prcc_my/{tardir}'

    os.makedirs(tar_dir, exist_ok=True)
    for idx in  tqdm(os.listdir(dir)):
        subdir = osj(dir,idx)
        for file in tqdm(os.listdir(subdir)):
            name,ext = os.path.splitext(file)
            name_split = name.split('_')
            if name_split[0]=='A':
                camid = '00'
                cloid = '01'
            elif name_split[0]=='B':
                camid = '01'
                cloid = '01'
            elif name_split[0]=='C':
                camid = '02'
                cloid = '02'
            random_frame = name_split[-1][-3:] # 001
            # print(random_frame)
            
            # '{:04d}_'.format(idx) 
            # save_name = f'{int(idx):04d}' + '_' + cloid + '_' + camid + '_' + random_frame + ext
            save_name = f'{int(idx):04d}' + '_' + f'{int(camid):02d}' + '_' + f'{int(cloid):02d}' + '_' + random_frame + ext
            save_path = osj(tar_dir,save_name)
            # print(save_path)
            shutil.copy(osj(subdir,file), save_path)
        #     break
        # break
```
